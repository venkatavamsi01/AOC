```python
def run_guard_simulation(grid):
    rows = len(grid)
    cols = len(grid[0]) if rows > 0 else 0

    directions = {
        '^': (-1, 0),
        '>': (0, 1),
        'v': (1, 0),
        '<': (0, -1)
    }
    order = ['^', '>', 'v', '<']

    def turn_right(d):
        i = order.index(d)
        return order[(i + 1) % 4]
    start_r, start_c, start_dir = None, None, None
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] in directions.keys():
                start_r, start_c = r, c
                start_dir = grid[r][c]
                break
        if start_r is not None:
            break
    cur_r, cur_c = start_r, start_c
    cur_dir = start_dir
    visited = set()

    while True:
        state = (cur_r, cur_c, cur_dir)
        if state in visited:
            return False
        visited.add(state)

        dr, dc = directions[cur_dir]
        front_r = cur_r + dr
        front_c = cur_c + dc
        if 0 <= front_r < rows and 0 <= front_c < cols:
            if grid[front_r][front_c] == '#':
                cur_dir = turn_right(cur_dir)
            else:
                cur_r, cur_c = front_r, front_c
        else:
            return True


def new_obstacle(original_grid):
    rows = len(original_grid)
    cols = len(original_grid[0]) if rows > 0 else 0
    guard_symbol = None
    for r in range(rows):
        for c in range(cols):
            if original_grid[r][c] in ['^','>','v','<']:
                guard_symbol = original_grid[r][c]
                break
        if guard_symbol is not None:
            break

    count = 0
    for r in range(rows):
        for c in range(cols):
            if original_grid[r][c] == '.':
                new_grid = [row[:] for row in original_grid]
                new_grid[r][c] = '#'
                leaves = run_guard_simulation(new_grid)
                if not leaves:
                    count += 1

    return count

result = new_obstacle(Orig_grid)
print(result)
```
